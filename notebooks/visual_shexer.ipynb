```python
from rdflib import Graph
from shexer.shaper import Shaper
from shexer.consts import NT, TURTLE
from shexviz import shex2dot
```

```python
rdf = Graph()
rdf.parse("http://www.wikidata.org/entity/Q42.ttl")
```

```python
query = """
SELECT * WHERE {
   ?boat wdt:P31 wd:Q35872 .
}
LIMIT 10

"""



target_classes = []
x = rdf.query(q)
for target_class in x:
  if str(target_class["class"]) not in target_classes:
    target_classes.append(str(target_class["class"]))

shex_target_file = "shexer.shex"

shaper = Shaper(target_classes=target_classes,
                rdflib_graph=rdf,
                input_format=TURTLE,
                instantiation_property="http://www.wikidata.org/prop/direct/P31"
                )  # Default rdf:type

shaper.shex_graph(output_file=shex_target_file)
```

```python
f = open("shexer.shex", "r")
shex = f.read()
gv = shex2dot.shex2dot(shex, "test")
gv.view()
```
